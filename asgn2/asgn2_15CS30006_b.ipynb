```python
from collections import defaultdict
def get_top_vocab_words(data,topK =  2000):
    vocab = defaultdict(int)
    for line in data:
        for token in line[1]:
            vocab[token]+=1
            
    sorted_vocab = sorted(vocab.items(),key=lambda x: x[1],reverse=True)
    return sorted_vocab[:topK]
```

```python
def make_indexed_map(data):
    ans = {}
    i = 0
    for line in data:
        ans[line[0]] = i
        i+=1
    return ans
```

```python
VECTOR_SIZE = 2000
myvocab = make_indexed_map(get_top_vocab_words(raw_data,VECTOR_SIZE))
myvocab
```

```python
THRESHOLD = 0.5
THRESHOLD_TANH = 0
def sigmoid(x):
    return 1/(1+ np.exp(-x))

def classify(x,thresh = 0.5):
    for line in x:
        if line[0] > thresh:
            line[0] = 1
        else:
            line[0] = 0
    return x

def classify_tanh(x, thresh = 0):
    for line in x:
        if line[0] > thresh:
            line[0] = 1
        else:
            line[0] = 0
    return x

def squared_error_func(predicted,truth):
    #return np.sum(truth*np.square(1 - predicted) + (1-truth)*np.square(predicted))
    return np.mean(np.square(truth - predicted))
        
def cross_entropy(predicted,truth):
    return -np.mean(truth*np.log(predicted) + (1-truth)*np.log(1 - predicted))

def classify_softmax(prob):
    result = np.zeros((prob.shape[0],1))
    i=0
    for line in prob:
        if line[1] >= line[0]:
            result[i,0] = 1
        i+=1
    return result

encoded_data.shape
```

```python
np.random.seed(0)
class NeuralNet:
    def __init__(self,X,y,hidden_1,hidden_2,out,activation="sigmoid"):
        self.X = X
        self.y = y
        self.input_dim = X.shape[1]
        self.num_expamples = X.shape[0]
        self.hidden_1 = hidden_1
        self.hidden_2 = hidden_2
        self.out_dim = out
        self.activation = activation
        self.layer1 = np.random.randn(self.input_dim,self.hidden_1) 
        self.hiddenLayer1 = np.random.randn(self.hidden_1, self.hidden_2) 
        self.hiddenLayer2 = np.random.randn(self.hidden_2,self.out_dim) 
        self.layer1_bias = np.zeros(self.hidden_1)
        self.hiddenLayer1_bias = np.zeros(self.hidden_2)
        self.hiddenLayer2_bias = np.zeros(self.out_dim)
    
    def train(self, lr=0.1,reg=0):
        #Forward pass
        out1 = self.X.dot(self.layer1) + self.layer1_bias
        #print(out1.shape)
        if self.activation == "sigmoid":
            z1 = sigmoid(out1)
        elif self.activation == "tanh":
            z1 = np.tanh(out1)
        #print(z1.shape)
        out2 = z1.dot(self.hiddenLayer1) + self.hiddenLayer1_bias
        #print(out2.shape)
        if self.activation == "sigmoid":
            z2 = sigmoid(out2)
        elif self.activation == "tanh":
            z2 = np.tanh(out2)
        #print(z2.shape)
        out = z2.dot(self.hiddenLayer2) + self.hiddenLayer2_bias
        
        nums = np.exp(out)
        dem = np.sum(nums,axis=1)
        dem = dem.reshape((dem.shape[0],1))
        #z = nums/dem
        #print("z ",z.shape)
        
        #classification = classify(z,THRESHOLD)
        #loss = squared_error_func(z, self.y)
        #loss = cross_entropy(z, self.y)
        # -ve loss 
        ylabels = [x[0] for x in self.y]
        loss = -np.sum(out[np.arange(self.num_expamples),ylabels])
        loss += np.sum(np.log(dem))
        loss /= self.num_expamples
        #print(ylabels)
        #loss = np.mean(-np.log(z[np.arange(self.num_expamples),ylabels]))
        print("Cost",loss)
        
        #Backward pass
        #dz = 2*(z - self.y)/self.num_expamples #dL/dz
        #dz = (z - self.y)/(z*(1-z)*self.num_expamples) #dL/dz
        #print("dL/dz\t",dz)
        #if self.activation == "sigmoid":
        #dout = dz*z*(1-z) #dL/dout
        dout = nums/dem
        dout[np.arange(self.num_expamples),ylabels] -= 1 
        dout /= self.num_expamples
        dhiddenLayer2_bias = np.sum(dout,axis=0)
        dhiddenLayer2 = z2.T.dot(dout)
        dz2 = dout.dot(self.hiddenLayer2.T)
        if self.activation == "sigmoid":
            dout2 = dz2*z2*(1-z2)
        elif self.activation == "tanh":
            dout2 = dz2*(1 - np.square(z2))
        dhiddenLayer1_bias = np.sum(dout2,axis=0)
        dhiddenLayer1 = z1.T.dot(dout2)
        dz1 = dout2.dot(self.hiddenLayer1.T)
        if self.activation == "sigmoid":
            dout1 = dz1*z1*(1-z1)
        elif self.activation == "tanh":
            dout1 = dz1*(1 - np.square(z1))
        dlayer1_bias = np.sum(dout1,axis=0)
        dlayer1 = self.X.T.dot(dout1)
        
        #Update
        self.layer1 -= lr*dlayer1
        self.layer1_bias -= lr*dlayer1_bias
        self.hiddenLayer1 -= lr*dhiddenLayer1
        self.hiddenLayer1_bias -= lr*dhiddenLayer1_bias
        self.hiddenLayer2 -= lr*dhiddenLayer2
        self.hiddenLayer2_bias -= lr*dhiddenLayer2_bias  
        
    def predict(self,X):
        out1 = X.dot(self.layer1) + self.layer1_bias
        if self.activation == "sigmoid":
            z1 = sigmoid(out1)
        elif self.activation == "tanh":
            z1 = np.tanh(out1)
        #print(z1.shape)
        out2 = z1.dot(self.hiddenLayer1) + self.hiddenLayer1_bias
        #print(out2.shape)
        if self.activation == "sigmoid":
            z2 = sigmoid(out2)
        elif self.activation == "tanh":
            z2 = np.tanh(out2)
        #print(z2.shape)
        out = z2.dot(self.hiddenLayer2) + self.hiddenLayer2_bias
        #print(out.shape)
        #if self.activation == "sigmoid":
        #z = sigmoid(out)
        #elif self.activation == "tanh":
        #    z = np.tanh(out)
        nums = np.exp(out)
        dem = np.sum(nums,axis=1)
        dem = dem.reshape((dem.shape[0],1))
        z = nums/dem
        #if self.activation == "sigmoid": 
        return classify_softmax(z)
        #elif self.activation == "tanh":
        #    return classify_tanh(z, THRESHOLD)
```

```python
labels = np.array([x[0] for x in raw_data]).reshape((encoded_data.shape[0],1))
train_examples = int(0.8*encoded_data.shape[0])
test_examples = encoded_data.shape[0] - train_examples
train_examples, test_examples
```

```python
np.random.seed(2)
permuatation = np.random.permutation(encoded_data.shape[0])
shuffled_data = encoded_data[permuatation]
shuffled_labels = labels[permuatation]
```

```python
X_train = shuffled_data[:train_examples]
y_train = shuffled_labels[:train_examples]
X_test = shuffled_data[train_examples:]
y_test = shuffled_labels[train_examples:]
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
HIDDEN_LAYER_1 = 100
HIDDEN_LAYER_2 = 50
OUTPUT_LAYER = 2
nn = NeuralNet(X_train, y_train, HIDDEN_LAYER_1, HIDDEN_LAYER_2, OUTPUT_LAYER)
IN_SAMPLE_ERROR = []
OUT_SAMPLE_ERROR = []
```

```python
for i in range(4000):
    train_pred = nn.predict(X_train)
    test_pred = nn.predict(X_test)
    in_sample_error = 1 - np.sum(train_pred == y_train)/y_train.shape[0]
    out_sample_error = 1 - np.sum(test_pred == y_test)/y_test.shape[0]
    IN_SAMPLE_ERROR.append(in_sample_error)
    print("Epoch: ",i)
    OUT_SAMPLE_ERROR.append(out_sample_error)
    print("In Sample Error:",in_sample_error)
    print("Out of Sample Error: ",out_sample_error)
    nn.train(0.5)
    
```
