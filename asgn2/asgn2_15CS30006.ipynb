```python
def get_vocab_count(data):
    vocab = set()
    for line in data:
        for token in line[1]:
            vocab.add(token)
    return len(vocab)
```

```python
from collections import defaultdict
def get_top_vocab_words(data,topK =  2000):
    vocab = defaultdict(int)
    for line in data:
        for token in line[1]:
            vocab[token]+=1
            
    sorted_vocab = sorted(vocab.items(),key=lambda x: x[1],reverse=True)
    return sorted_vocab[:topK]
```

```python
def make_indexed_map(data):
    ans = {}
    i = 0
    for line in data:
        ans[line[0]] = i
        i+=1
    return ans
```

```python
VECTOR_SIZE = 2000
myvocab = make_indexed_map(get_top_vocab_words(raw_data,VECTOR_SIZE))
myvocab
```

```python
encoded_data = np.zeros((len(raw_data),len(myvocab)))
i=0
for line in raw_data:
    for token in line[1]:
        if token in myvocab:
            encoded_data[i,myvocab[token]] = 1
    i+=1
encoded_data
```

```python
THRESHOLD = 0.5
def sigmoid(x):
    return 1/(1+ np.exp(-x))

def classify(x,thresh = 0.5):
    for line in x:
        if line[0] > thresh:
            line[0] = 1
        else:
            line[0] = 0
    return x

def squared_error_func(predicted,truth):
    return np.sum(truth*np.square(1 - predicted) + (1-truth)*np.square(predicted))
        

encoded_data.shape
```

```python
class NeuralNet:
    def __init__(self,X,y,hidden_1,hidden_2,out,activation="sigmoid"):
        self.X = X
        self.y = y
        self.input_dim = X.shape[1]
        self.num_expamples = X.shape[0]
        self.hidden_1 = hidden_1
        self.hidden_2 = hidden_2
        self.out_dim = out
        self.activation = "sigmoid"
        self.layer1 = np.random.rand(self.input_dim,self.hidden_1)
        self.hiddenLayer1 = np.random.rand(self.hidden_1, self.hidden_2)
        self.hiddenLayer2 = np.random.rand(self.hidden_2,self.out_dim)
        self.layer1_bias = np.zeros(self.hidden_1)
        self.hiddenLayer1_bias = np.zeros(self.hidden_2)
        self.hiddenLayer2_bias = np.zeros(self.out_dim)
    
    def train(self, lr=0.1,reg=0):
        #Forward pass
        out1 = self.X.dot(self.layer1) + self.layer1_bias
        print(out1.shape)
        if self.activation == "sigmoid":
            z1 = sigmoid(out1)
        elif self.activation == "tanh":
            z1 = np.tanh(out1)
        print(z1.shape)
        out2 = z1.dot(self.hiddenLayer1) + self.hiddenLayer1_bias
        print(out2.shape)
        if self.activation == "sigmoid":
            z2 = sigmoid(out2)
        elif self.activation == "tanh":
            z2 = np.tanh(out2)
        print(z2.shape)
        out = z2.dot(self.hiddenLayer2) + self.hiddenLayer2_bias
        print(out.shape)
        if self.activation == "sigmoid":
            z = sigmoid(out)
        elif self.activation == "tanh":
            z = np.tanh(out)
        print(z.shape)
        
        #classification = classify(z,THRESHOLD)
        loss = squared_error_func(z, self.y)
        print(loss)
        
        #Backward pass
        dz = 2*(2*z - self.y*z - 1) #dL/dz
        print("dL/dz\t",dz.shape)
        if self.activation == "sigmoid":
            dout = dz*out*(1-out) #dL/dout
        print("dL/dout\t",dout.shape)
        dhiddenLayer2_bias = np.sum(dout,axis=0)
        #dhiddenLayer2 = np.sum(z2.T.dot(dout),axis=1)
        dhiddenLayer2 = z2.T.dot(dout)
        dhiddenLayer2_bias = dhiddenLayer2_bias.reshape((dhiddenLayer2_bias.shape[0],1))
        #dhiddenLayer2 = dhiddenLayer2.reshape((dhiddenLayer2.shape[0],1))
        print("dL/dbh2\t",dhiddenLayer2_bias.shape)
        print("dL/dwh2\t",dhiddenLayer2.shape)
        dz2 = dout.dot(dhiddenLayer2.T)
        print("dL/dz2\t",dz2.shape)
        if self.activation == "sigmoid":
            dout2 = dz2*out2*(1-out2)
        print("dL/dout2\t",dout2.shape)
        dhiddenLayer1_bias = np.sum(dout2,axis=0)
        #dhiddenLayer1 = np.sum(z1.T.dot(dout2),axis=1)
        dhiddenLayer1 = z1.T.dot(dout2)
        dhiddenLayer1_bias = dhiddenLayer1_bias.reshape((dhiddenLayer1_bias.shape[0],1))
        #dhiddenLayer1 = dhiddenLayer1.reshape((dhiddenLayer1.shape[0],1))
        print("dL/dbh1\t",dhiddenLayer1_bias.shape)
        print("dL/dwh1\t",dhiddenLayer1.shape)
        dz1 = dout2.dot(dhiddenLayer1.T)
        print("dL/dz1\t",dout2.shape)
        if self.activation == "sigmoid":
            dout1 = dz1*out1*(1-out1)
        print("dL/dout1\t",dout1.shape)
        dlayer1_bias = np.sum(dout1,axis=0)
        #dhiddenLayer1 = np.sum(z1.T.dot(dout2),axis=1)
        dlayer1 = self.X.T.dot(dout1)
        dlayer1_bias = dlayer1_bias.reshape((dlayer1_bias.shape[0],1))
        #dhiddenLayer1 = dhiddenLayer1.reshape((dhiddenLayer1.shape[0],1))
        print("dL/db\t",dlayer1.shape)
        print("dL/dw\t",dlayer1_bias.shape)
        
        
```
